ოდნავ გავცდეთ მონეტებს, კამათლებს და ბანქოს და წარმოვიდგინოთ, რომ ფეხბურთს უყურებთ

Ასევე, წარმოვიდგინოთ, რომ ფეხბურთის არაფერი გაგეგებათ, და მხოლოდ ის იცით, რომ ან წითელმა გუნმდა უნდა გაიმარჯვოს ან ლურჯმა გუნდმა (ფრე ვერ დამთავრდება); რა იქნებოდა თქვენთვის წითელი გუნდის გამარჯვების ალბათობა მხოლოდ ამ ინფორმაციით?

```python
def predict_football(red_team_info, blue_team_info, time_left, current_score):
    print(f"წითელი გუნდის ინფო: {red_team_info}")
    print(f"ლურჯი გუნდის ინფო: {blue_team_info}")
    print(f"დარჩენილი დრო: {time_left} წუთი")
    print(f"მიმდინარე ანგარიში: {current_score}")

    if red_team_info == "მსოფლიო ჩემპიონი":
        prob = 0.6
    elif blue_team_info == "ნაკლული შემადგენლობა":
        prob = 0.55
    else:
        prob = 0.5

    if time_left <= 5 and "ლურჯი იგებს" in current_score:
        prob = 0.3

    print(f"\nწითელი გუნდის გამარჯვების ალბათობა: {prob:.2f}")

interactive_prediction = interactive(
    predict_football,
    red_team_info=widgets.Dropdown(options=["ჩვეულებრივი", "მსოფლიო ჩემპიონი"], description="წითელი გუნდი:"),
    blue_team_info=widgets.Dropdown(options=["ჩვეულებრივი", "ნაკლული შემადგენლობა"], description="ლურჯი გუნდი:"),
    time_left=widgets.IntSlider(min=0, max=90, step=5, value=90, description="დარჩენილი დრო:"),
    current_score=widgets.Dropdown(options=["0-0", "წითელი იგებს", "ლურჯი იგებს"], description="მიმდინარე ანგარიში:")
)

display(interactive_prediction)
```

**დიდ რიცხვთა კანონი** - ეს კანონი საფუძვლად უდევს მათემატიკურ სტატისტიკას. დიდ რიცხვთა კანონის თანახმად, **შერჩევითი მახასიათებლები ცდების (ექსპერიმენტების) რიცხვთა ზრდისას უახლოვდება თეოირიულ მახასიათებელს** , რაც საშუალებას იძლევა, ამა თუ იმ რეალური მოვლენის ალბათური მოდელების პარამეტრები შევაფასოთ ცდების მიერ მიღებული შედეგების გამოყენებით.

ცხადია, ჩვენთვის ცნობილი თეორიული ალბათობებიც უამრავი დაკვირვების და ექსპერიმენტის შედეგია. მაგ. მე-18 საუკუნეში ბუნებისმეტყველმა ბიუფონმა მონეტა ააგდო 4040-ჯერ, საიდანაც გერბი მოვიდა 2048-ჯერ, მე-20 საუკუნის დასაწყისში მათემატიკოსმა პირსონმა მონეტა ააგდო 24 000-ჯერ, საიდანაც გერბი მოვიდა 12 012-ჯერ. შესაბამისად, მიუხედავად იმისა, რომ მონეტის მრავალჯერადი აგდების შედეგები შემთხვევითი ხდომილებაა, მონეტის მრავალჯერადი აგდების შედეგები ობიექტურ კანონს ემორჩილება.

```python
def coin_flip_simulation(num_flips):
    flips = np.random.choice(['თავი', 'საფასური'], size=num_flips)
    heads_ratio = np.cumsum(flips == 'თავი') / np.arange(1, num_flips + 1)

    plt.figure(figsize=(10, 6))
    plt.plot(range(1, num_flips + 1), heads_ratio)
    plt.axhline(y=0.5, color='r', linestyle='--')
    plt.xlabel('აგდების რაოდენობა')
    plt.ylabel('თავის მოსვლის პროპორცია')
    plt.title('მონეტის აგდების სიმულაცია')
    plt.show()

interactive_coin_flip = interactive(coin_flip_simulation, num_flips=widgets.IntSlider(min=10, max=10000, step=10, value=100, description='აგდების რაოდენობა:'))
display(interactive_coin_flip)
```

მარტივი მაგალითისთვის, წარმოვიდგინოთ, რომ გვინდა დავადგინოთ, ხვიჩა კვარაცხელია გოლს გაიტანს თუ არა შემდეგ მატჩში.

```python
def predict_kvara_goal(games_played, goals_scored):
    probability = goals_scored / games_played
    print(f"\nბოლო {games_played} თამაშში, ხვიჩამ {goals_scored} გოლი გაიტანა.")
    print(f"გოლის გატანის ალბათობა შემდეგ მატჩში: {probability:.2f}")

    plt.figure(figsize=(8, 6))
    plt.bar(['გოლი', 'არა გოლი'], [probability, 1-probability])
    plt.ylabel('ალბათობა')
    plt.title('ხვიჩა კვარაცხელიას გოლის გატანის ალბათობა')
    plt.show()

interactive_kvara = interactive(
    predict_kvara_goal,
    games_played=widgets.IntSlider(min=1, max=50, step=1, value=20, description='ნათამაშები მატჩები:'),
    goals_scored=widgets.IntSlider(min=0, max=50, step=1, value=15, description='გატანილი გოლები:')
)
display(interactive_kvara)
```

სასარგებლო იქნება თუ არა, Რომ ჩავატაროთ კომპიუტერული სიმულაცია, სადაც ორივე გუნდს ექნება გატანის თანაბარი ალბათობა ყოველ წუთს და საბოლოო ანგარიში 90 წუთის შემდეგ მიიღება? რამდენად დაგვეხმარება ასეთი სიმულაციის ბევრჯერ გაშვება რეალისტური პროგნოზის გასაკეთებლად?


მსგავსი სიმულაცია არ იქნება სასარგებლო, სულ მცირე Იმიტომ, რომ გატანის ალბათობა თანაბარი არ არის. არსებობს სხვადასხვა ფაქტორები, რომელიც შესაძლოა, ამ ალბათობას განსაზღვრავდეს და, ასევე, ცვლიდეს ყოველი მომენტისთვის  - შესაბამისად, სულ მცირე, გვჭირდება დამატებითი ინფორმაცია - ანუ მონაცემები, რომლის მიხედვითაც შევძლებთ ამ ალბათობის კორექტირებას ((Მაგალითად საქართველოს თუ გატანის ალბათობას 1ზე “დავუყენებთ” და ესპანეთს 0-ზე სულ საქართველო მოიგებს ყველა სიმულაციას)). ერთ-ერთი გზა ალბათობის გამოსათვლელად იქნებოდა გასული თამაშების მიხედვით გოლის გატანის ალბათობის გამოთვლა, თუმცა ეს იგივე იქნებოდა, რომ პირდაპირ გასული 7 თამაშის შედეგის მიხედვით გამოგვეთვალა შემდეგი მატჩის პროგნოზი.

 Რა უნდა ვქნათ რომ ჩვენი სიმულაცია და პროგნოზი მაქსიმალურად მივაახლოვოთ რეალობასთან?


გვიშველის თუ არა აქ რაიმე კომპლექსური მათემატიკური მოდელირება, ან ზემოთ აღწერილი სიმულაციის ბევრჯერ ჩატარება?


სინამდვილეში, როდესაც მხოლოდ ეს ინფორმაცია გვაქვს, ვერანაირი კომპლექსური მათემატიკა და ვერც მილიონი, მილიარდი სიმულაცია ვერ გვიშველის - რადგან არსობრივად არაფერი შეიცვლება, და ჩვენი მთელი პროგნოზი იმ გამოთვლილ ალბათობას დაეყრდნობა, რომელიც წარსული მატჩების გატანილი გოლების რაოდენობის მიხედვით დავითვალეთ - ანუ მხოლოდ ეს იქნება ინფორმაციის წყარო, რომლითაც თუნდაც სიმულაციის მოდელი იხელმძღვანელებს.

 “გამოსავალი”/პროგნოზის გაუმჯობესების გზა მონაცემების უფრო დეტალურ ანალიზშია - ისეთი ფაქტორების, როგორიცაა მაგალითად, ფეხბურთელების ტრამვები, ამინდი, გუნდის შემადგენლობა და სხვა მრავალი ფაქტორი, რომელიც სინამდვილეში გავლენას ახდენს მატჩის შედეგზე. იმ გზებს, თუ როგორ შეიძლება მსგავსი ინფორმაციები მოვიძიოთ, დავამუშავოთ, და პროგნოზირებისთვის გამოვიყენოთ ჩვენს კურსში მოგვიანებით გავეცნობით.
