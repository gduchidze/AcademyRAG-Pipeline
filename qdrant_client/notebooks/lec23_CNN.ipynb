# კონვოლუციის მარტივი მაგალითები

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
input_matrix_1 = np.ones((10,5))*150
```

```python
input_matrix_2 = np.ones((10,5))*0
```

```python
image = np.concatenate((input_matrix_1, input_matrix_2), axis=1)
```

```python
image
```

```python
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```

### ავიღოთ მარტივი ფილტრი რომელიც კუთხეების ამოცნობაში დაგვეხმარება

```python
filter_edge = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
filter_edge
```

```python
from scipy.signal import convolve2d
```

```python
conv_image = convolve2d(image, filter_edge,mode='valid')
conv_image
```

```python
plt.imshow(conv_image, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```
