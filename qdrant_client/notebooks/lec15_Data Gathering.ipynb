# ლექცია 2

## მონაცემთა მოპოვება - შესავალი


## რატომ არის მონაცემები გადამწყვეტი ML & NLP-ში
მანქანური სწავლების (ML) და ბუნებრივი ენის დამუშავების (NLP) მოდელების პერფორმანსი მნიშვნელოვნად არის დამოკიდებული მონაცემებზე. მაღალი ხარისხის, შესაბამისი მონაცემები აუცილებელია მოდელების გასაწვრთნელად, რომლებსაც რთული შაბლონების გაგება და მათი პროგნოზირება შეუძლიათ. შესაბამისად, მონაცემთა მოპოვება ერთ-ერთი ყველაზე მნიშვნელოვანი და გადამწყვეტი უნარია მონაცემთა მეცნიერების ნებისმიერ პროექტში.

### 1. მონაცემთა მნიშვნელობა ML & NLP-ში

მანქანური სწავლების (ML) და ბუნებრივი ენის დამუშავების (NLP) მოდელების პერფორმანსი მნიშვნელოვნად არის დამოკიდებული მონაცემებზე. მაღალი ხარისხის, შესაბამისი მონაცემები აუცილებელია მოდელების გასაწვრთნელად, რომლებსაც რთული შაბლონების გაგება და მათი პროგნოზირება შეუძლიათ. შესაბამისად, მონაცემთა მოპოვება ერთ-ერთი ყველაზე მნიშვნელოვანი და გადამწყვეტი უნარია მონაცემთა მეცნიერების ნებისმიერ პროექტში.

#### 1.1 მოდელის ეფექტურობა
- მაღალი ხარისხის მონაცემები უზრუნველყოფს მოდელის უკეთეს სწავლებას და შედეგად, უფრო ზუსტ პროგნოზებს.
- არასაკმარისი ან დაბალი ხარისხის მონაცემები იწვევს ცუდად განზოგადებად მოდელებს.

#### 1.2 პრობლემის გადაჭრის უნარი
- მრავალფეროვანი და რეპრეზენტატული მონაცემები ეხმარება მოდელს რეალური სამყაროს პრობლემების უკეთ გადაჭრაში.
- კარგი მონაცემები საშუალებას აძლევს მოდელს გაიგოს კონტექსტი და ნიუანსები.

#### 1.3 ინოვაცია და აღმოჩენები
- ახალი და მრავალფეროვანი მონაცემები ხელს უწყობს ინოვაციურ გადაწყვეტილებებს ML და NLP-ში.
- დიდი მოცულობის მონაცემები შეიძლება გამოყენებულ იქნას ახალი შაბლონებისა და ტენდენციების აღმოსაჩენად.

### 2. Data Pipeline

"Data Pipeline" არის პროცესების სერია, რომლის განმავლობაშიც გროვდება, სუფთავდება, მუშავდება და მზადდება მონაცემები შემდგომი გამოყენებისთვის. აღნიშნული პროცესი უზრუნველყოფს მონაცემების მაღალ ხარისხს და ამოცანისთვის შესაბამისობას, რასაც შემდგომში უკვე დიდი გავლენა აქვს შედეგების საიმედოობასა და სიზუსტეზე. 

#### 2.1 Data Pipeline-ის ეტაპები:
1. **შეგროვება:** 
   - მონაცემების შეგროვება სხვადასხვა წყაროებიდან (მონაცემთა ბაზები, API-ები, ფაილები).
   - წყაროების იდენტიფიკაცია და მონაცემთა ექსტრაქციის მეთოდების შერჩევა.

2. **გასუფთავება:** 
   - მონაცემებში შეცდომებისა და შეუსაბამობების აღმოფხვრა ან გასწორება.
   - დუბლიკატების მოშორება, ხარვეზების შევსება, ფორმატირების სტანდარტიზაცია.

3. **წინასწარი დამუშავება:** 
   - ნედლი მონაცემების გარდაქმნა მოდელის სწავლებისთვის/წვრთნისთვის შესაფერის ფორმატში.
   - ფიჩერების ინჟინერია, ნორმალიზაცია, კოდირება.

4. **ვალიდაცია:**
   - მონაცემთა ხარისხისა და შესაბამისობის შემოწმება.
   - სტატისტიკური ანალიზი და ანომალიების დეტექცია.

5. **შენახვა:**
   - დამუშავებული მონაცემების უსაფრთხო და ეფექტური შენახვა.
   - მონაცემთა ბაზების ან ღრუბლოვანი სერვისების გამოყენება.

6. **ტრანსფორმაცია:**
   - მონაცემთა სტრუქტურის ან ფორმატის შეცვლა კონკრეტული ანალიზისთვის.
   - აგრეგაცია, ჯოინები, პივოტირება.

7. **ანალიზი და ვიზუალიზაცია:**
   - მონაცემთა ანალიზი და ვიზუალური რეპრეზენტაცია.
   - სტატისტიკური მეთოდების და ვიზუალიზაციის ინსტრუმენტების გამოყენება.

### 3. მონაცემთა ტიპები

მანქანური სწავლების პროცესში სხვადასხვა ტიპის მონაცემები გამოიყენება. თითოეულ მათგანს ჭირდება შესაბამისი დამუშავება შემდეგი ეტაპებისთვის გასამზადებლად. 

#### 3.1 სტრუქტურირებული მონაცემები:
- ორგანიზებულია ფიქსირებულ ფორმატში (ცხრილები, მონაცემთა ბაზები).
- ადვილად ანალიზირებადი და დასამუშავებელი.
- მაგალითები: Excel ცხრილები, SQL მონაცემთა ბაზები.

#### 3.2 არასტრუქტურირებული მონაცემები:
- არ გააჩნია წინასწარ განსაზღვრული სტრუქტურა.
- მოიცავს ტექსტს, სურათებს, ვიდეოებს, ხმოვან ჩანაწერებს.
- მაგალითები: სოციალური მედიის პოსტები, ელ-ფოსტის შეტყობინებები.

#### 3.3 ნახევრად სტრუქტურირებული მონაცემები:
- შეიცავს როგორც სტრუქტურირებული, ასევე არასტრუქტურირებული მონაცემების ელემენტებს.
- მაგალითები: JSON და XML ფაილები, ელ-ფოსტები მეტა-მონაცემებით.

### 4. მაღალი ხარისხის მონაცემების კრიტიკული როლი მონაცემთა მეცნიერებაში

#### 4.1 რატომ არის მაღალი ხარისხის მონაცემები აუცილებელი:

1. **სიზუსტე:** 
   - უკეთესი ხარისხის მონაცემებით შეგვიძლია შედარებით ზუსტი მოდელის მიღება.
   - ამცირებს ცრუ დადებითი (False Positive) და უარყოფითი (False Negative) შედეგების რაოდენობას.

2. **სიმტკიცე:** 
   - საშუალებას გვაძლევს ავაგოთ უფრო ძლიერი/"შეუვალი" მოდელები.
   - მოდელები უკეთ უმკლავდებიან ახალ, უცნობ მონაცემებს.

3. **შესაბამისობა:** 
   - კონკრეტულ სფეროზე მორგებული მონაცემები ეხმარება მოდელს კონტექსტის გაგებაში.
   - უზრუნველყოფს მოდელის რელევანტურობას კონკრეტული ამოცანისთვის.

4. **განზოგადების უნარი:**
   - ხარისხიანი მონაცემები აუმჯობესებს მოდელის უნარს, სწორად იმუშაოს ახალ, უცნობ სიტუაციებში.

5. **ეფექტურობა რესურსების გამოყენებაში:**
   - მაღალი ხარისხის მონაცემები ამცირებს საჭირო დამუშავების დროს და კომპიუტერულ რესურსებს.

#### 4.2 მონაცემთა ხარისხის კონტროლი:

მაღალი ხარისხის მონაცემების მისაღებად აუცილებელია შემდეგი ეტაპების გავლა:

1. **მონაცემთა ვალიდაცია:**
   - შემოწმება სისრულეზე, თანმიმდევრულობასა და სიზუსტეზე.
   - ანომალიების და გამონაკლისების იდენტიფიკაცია.

2. **გაწმენდა:**
   - დუბლიკატების მოშორება.
   - ხარვეზების შევსება ან მოშორება.
   - ფორმატირების სტანდარტიზაცია.

3. **აუგმენტაცია:**
   - არსებული მონაცემების გამდიდრება დამატებითი წყაროებით.
   - ხელოვნური მონაცემების გენერირება ბალანსის დასაცავად.

4. **ნორმალიზაცია:**
   - მონაცემთა მასშტაბირება ერთიან დიაპაზონში.
   - კატეგორიული ცვლადების კოდირება.


#### 4.3 გავლენა მოდელის შესაძლებლობებზე:

მაღალი ხარისხის მონაცემები მნიშვნელოვნად აუმჯობესებს მოდელის შესაძლებლობებს:

1. **გაუმჯობესებული პროგნოზირების სიზუსტე:**
   - უფრო ზუსტი და სანდო შედეგები (როგორც ნაცნობ, ასევე ახალ, უცნობ მონაცემებზე.)

2. **ნაკლები მიკერძოება:**
   - ამცირებს სისტემურ შეცდომებს და უსამართლო დისკრიმინაციას.

3. **სწრაფი სწავლება:**
   - ხარისხიანი მონაცემები აჩქარებს მოდელის სწავლების პროცესს.

ცუდი ხარისხის მონაცემებმა, პირიქით, შეიძლება შეცდომაში შეიყვანოს მოდელები და არასასურველი შედეგი მივიღოთ, რაც გამოიხატება დაბალ სიზუსტეში, მიკერძოებულ შედეგებში და არასანდო პროგნოზებში.

### 5. Მონაცემთა წყაროები: Kaggle, Wikipedia, Huggingface, API, PDFs, scraping...

მონაცემთა მეცნიერებისა და მანქანური სწავლების სფეროში, მონაცემთა წყაროების ხარისხი და მრავალფეროვნება კრიტიკულ როლს თამაშობს ნებისმიერი პროექტის წარმატებაში. ქვემოთ მოცემულია რამდენიმე ძირითადი მონაცემთა წყარო, რომელთა გამოყენებაც თავისუფლადაა შესაძლებელი:

##### Kaggle

Kaggle ერთ-ერთი ყველაზე პოპულარული პლატფორმაა მონაცემთა მეცნიერების სფეროში. გარდა იმისა, რომ ის ცნობილია თავისი კონკურენტული გარემოთი და მრავალფეროვანი მონაცემთა ნაკრებებით, Kaggle ასევე გამოირჩევა დიდი რაოდენობის სასწავლო რესურსებით. პლატფორმაზე ხელმისაწვდომია ათასობით სასწავლო ნოუთბუქი, რომლებიც მოიცავს თემებს დაწყებული მონაცემთა ანალიზის საფუძვლებიდან, მანქანური სწავლების ალგორითმებით და დამთავრებული რთული ხელოვნური ინტელექტის მოდელებით.

შემდეგ ეტაპზე ვახდენთ Wikipedia API-ს ინიციალიზაციას ქართული ენისთვის:

````python
import wikipediaapi

# გავამზადოთ Wikipedia API ქართული  ვიკიპედიისთვის
# საჭიროა შემდეგი ფორმატი: wikipediaapi.Wikipedia('სახელი (email)', 'ენა')
wiki_wiki = wikipediaapi.Wikipedia('MyWikipediProject (MyWikipediProject@gmail.com)', 'ka')
wiki_wiki
````

ამ კოდში:
- 'MyWikipediProject (MyWikipediProject@gmail.com)' არის იდენტიფიკატორი, რომელიც გამოიყენება Wikipedia-ს სერვერებზე მოთხოვნების გასაგზავნად, რაც Wikipedia-ს საშუალებას აძლევს თვალი ადევნოს API-ს გამოყენებას.
- 'ka' არის ენის კოდი ქართული ენისთვის (ყველა ენას გააჩნია უნიკალური კოდი).

##### გამოყენების მაგალითი

````python
# განვსაზღვროთ ფუნქცია რომელიც სათაურზე დაყრდნობით წამოიღებს სტატიის ტექსტს
def fetch_wikipedia_article(title: str) -> None:
    page = wiki_wiki.page(title)
    if page.exists():
        print(f"სათაური: {page.title}\n")
        print(f"სრული ტექსტი: {page.text[:1000]}...")  # დავბეჭდოთ მხოლოდ 1000 სიმბოლო
    else:
        print(f"გვერდი სახელად '{title}' არ არსებობს.")
````

````python
# გვერდის ურლ: https://ka.wikipedia.org/wiki/დავით_IV_აღმაშენებელი
article_title = "დავით IV აღმაშენებელი" # შეიძლება ნებისმიერი არსებული სტატიის ტექსტის წამოღება
fetch_wikipedia_article(article_title)
````

#### დავაკონვერტიროთ PDF-ები ტექსტში (დამატებითი)

მონაცემების კიდევ ერთი, საინტერესო საშუალებაა დოკუმენტებიდან (ამ შემთხვევაში, pdf-დან) მონაცემების მოპოვება. PDF (Portable Document Format) დოკუმენტების შესანახი ფართოდ გავრცელებული ფორმატია, ამიტომ ხშირად გვჭირდება ამ დოკუმენტებიდან ინფორმაციის ამოღება და დამუშავება. Python-ის დახმარებით შეგვიძლია ეს პროცესი გავამარტივოთ და ავტომატიზირებული გავხადოთ.

````python
# # !pip install PyPDF2 pycryptodome==3.15.0
from PyPDF2 import PdfReader

reader = PdfReader('pdfs/shota-rustaveli-vefxistyaosani.pdf')
page_texts = []
for page in reader.pages[:10]: # წავიკითხოთ pdf-ის პირველი 10 გვერდი 
    page_texts.append(page.extract_text())
text = '<new_page>'.join(page_texts)
print(text[:1000])
````
